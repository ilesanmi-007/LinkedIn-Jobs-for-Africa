```python
#from google.colab.output import eval_js
criteria_dict = df['criteria'].to_dict()
Job_function = [None] * len(criteria_dict)
Employment_status = [None] * len(criteria_dict)
Seniority_level = [None] * len(criteria_dict)
Industries = [None] * len(criteria_dict)


for i in criteria_dict:
  real_criteria_dict = eval(criteria_dict[i])

  S_level = list(map(lambda x:'Seniority level' in x.keys(), real_criteria_dict))
  E_status = list(map(lambda x:'Employment type' in x.keys(), real_criteria_dict))
  I_stries = list(map(lambda x:'Industries' in x.keys(), real_criteria_dict))
  J_func = list(map(lambda x:'Job function' in x.keys(), real_criteria_dict))

  if True in S_level:
    Seniority_level[i] = real_criteria_dict[S_level.index(True)]['Seniority level']
    #Seniority_level.append(S_level)
  if True in J_func:
    Job_function[i] = real_criteria_dict[J_func.index(True)]['Job function'] 
  if True in E_status:
    Employment_status[i] = real_criteria_dict[E_status.index(True)]['Employment type'] 
  if True in I_stries:
    Industries[i] = real_criteria_dict[I_stries.index(True)]['Industries'] 
  
  
```

```python
df['seniority level'] = Seniority_level
df['employment_status'] = Employment_status
df['industries'] = Industries
df['job_function'] = Job_function
df['ID'] = df.index
```

## **This below has all 'ands' removed:**

```python
  
# Create a new dataframe to store the industries for each ID
industries_df = pd.DataFrame(columns=['index', 'Industry'])

for index, row in df.iterrows():
    ID = row['ID']
    industries = row['industries']
    # check if the industries column is None
    if pd.isna(industries) or industries == 'None':
        industries_df = industries_df.append({'ID': ID, 'Industry': None}, ignore_index=True)
    else:
        # Replace 'and' with ','
        industries = industries.replace('and', ',')
        
        #industries = industries.apply(lambda x: re.sub(r"(?<!Information )and", ", ", x))
        if ',' in industries:
            # Iterate over the industries in the list
            for industry in industries.split(','):
                # Append a new row to the industries_df with the ID and industry
                industries_df = industries_df.append({'ID': ID, 'Industry': industry.strip()}, ignore_index=True)
        else:
            # Append a new row to the industries_df with the ID and industry
            industries_df = industries_df.append({'ID': ID, 'Industry': industries}, ignore_index=True)


# merge the industries_df with the original dataframe
df_new = pd.merge(df, industries_df, on='ID', how='left')
```
